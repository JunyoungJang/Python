```python
%matplotlib inline
import matplotlib.image as im
import matplotlib.pyplot as plt
from ipywidgets import Image
from IPython import display

animatedGif = "dcgan_ani.gif"
file = open(animatedGif , "rb")
image = file.read()
progress= Image(
    value=image,
    format='gif',
    width=250,
    height=250)
display.display(progress)
```

```python
vae_recon = 're_vae.png'
dcgan_recon = 're_dcgan.png'

fig, axes = plt.subplots(1, 2, figsize=(10, 10))

image=im.imread(vae_recon)
axes[0].imshow(image)
axes[0].set_ylabel('VAE', fontsize=50)

image=im.imread(dcgan_recon)
axes[1].imshow(image)
axes[1].set_ylabel('DCGAN', fontsize=50)

plt.tight_layout()
```
